***6. Creen un programa que permita dibujar diferentes tipos de figuras geométricas con el uso de eventos de mouse para poder hacer dibujos y luego guarden la imagen.***

```python
draw = False
imagen = np.zeros((600, 800, 3), np.uint8)

def draw(event,x,y,flags,param):
    global draw
    if event==cv2.EVENT_LBUTTONDOWN:
        draw=True
        cv2.circle(imagen, (x,y), 50 , (255,0,0), 2)

    
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(imagen, (x,y), (x+80, y+80), (0,255,0), 3)
```

```python

cv2.namedWindow('Imagen')
cv2.setMouseCallback('Imagen', draw)


while True:
    cv2.imshow('Imagen',imagen)
    if cv2.waitKey(1) & 0xFF == 27:
        break
    
cv2.destroyAllWindows()
```

***7. Modifiquen el programa***

```python
ix,iy,sx,sy = -1,-1,-1,-1
imagen = np.zeros((600, 800, 3), np.uint8)

def draw_lines(event, x, y, flags, param):
    global ix,iy,sx,sy
    
    if event == cv2.EVENT_LBUTTONDOWN:

        # draw circle of 2px
        cv2.circle(imagen, (x, y), 3, (0, 0, 127), -1)

        if ix != -1:
            cv2.line(imagen, (ix, iy), (x, y), (0, 0, 127), 2, cv2.LINE_AA)
        else: 
            sx, sy = x, y
            
        ix,iy = x, y
        
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        ix, iy = -1, -1 
        if flags == 33: 
            cv2.line(imagen, (x, y), (sx, sy), (0, 0, 127), 2, cv2.LINE_AA)
```

```python


cv2.namedWindow('image') 
cv2.setMouseCallback('image',draw_lines)

while(True):
    cv2.imshow('image',imagen)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
